### 9. Records of team who decided to bat first and won

### 11. Number of wins w.r.t each after batting first

### 12. Make barplot for top 5 teams with most wins after batting first

### 13. Make piechart for battingFirst with winning percentage

```python
plt.figure(figsize=(10,10))
plt.pie(list(battingFirst['winner'].value_counts()),labels=list(battingFirst['winner'].value_counts().keys()),autopct = '%0.1f%%')
plt.show()
```
